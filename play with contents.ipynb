```python
import os
import json
from collections import Counter, defaultdict
```

```python
udemy_data = list()
filenames = os.listdir("udemy_data")
for i, f in enumerate(filenames):
    course_data = json.load(open("udemy_data/" + f))
    udemy_data.append(course_data)
    #print("Done {}/{}".format(i+1, len(filenames)))
print("Done")
```

```python
all_titles_list = list()
first_titles_list = list()
second_titles_list = list()
third_titles_list = list()

for course in udemy_data:
    all_titles_list.append(course["t"].lower())
    first_titles_list.append(course["t"].lower())
    for c in course["c"]:
        all_titles_list.append(c["t"].lower())
        second_titles_list.append(c["t"].lower())
        for c2 in c["c"]:
            all_titles_list.append(c2["t"].lower())
            third_titles_list.append(c2["t"].lower())
```

```python
all_titles_counter = Counter(all_titles_list)
first_titles_counter = Counter(first_titles_list)
second_titles_counter = Counter(second_titles_list)
third_titles_counter = Counter(third_titles_list)
```

```python
second_titles_counter.most_common(200)
```

```python
#Get topics prereqs
topics = defaultdict(Counter)
for course in udemy_data: #Iterate thru every course
    course_contents = list() #List to register the current course contents
    for content in course["c"]: #Iterate thru content
        #For the current content, append every one that happened before it as pre reqs
        topics[content["t"].lower()].update(course_contents)
        course_contents.append(content["t"].lower())
```

```python
#topics["functions"].most_common()
#keep studying this

#subtract occurrences of the target knowledge from the pre reqs candidates
#Must take care with low occurrences contents because they will tend to stay at 1
target = "overview of c#"
prereqs_diff = Counter()
for prereq, _ in topics[target].most_common():
    
    prereq_count = topics[target][prereq] #Get the degree of need of the prereq in the target
    target_count = topics[prereq].get(target, 0) #Get the degree of need of the target in the prereq, return 0 if not found
    
    prereqs_diff[prereq] = prereq_count - target_count

print("\n".join(map(str, prereqs_diff.most_common())))
```

```python
courses_with_term = list()

for course in udemy_data:
    for c in course["c"]:
        if c["t"].lower() == "Classes":
            courses_with_term.append(course)

for course in courses_with_term:
    print(course["t"])
    for c in course["c"]:
        print("    " + c["t"])
    print("\n")
```

```python
def get_course_subsections(course_name):
    total_subsections = list()
    for course in udemy_data:
        if course["t"].lower() == course_name.lower():
            subsections = list()
            for subsec in course["c"]:
                subsections.append(subsec["t"])
            total_subsections.append(subsections)
    return total_subsections

same_courses_data = get_course_subsections("html for beginners")
for c in same_courses_data:
    for prereq in c:
        print(prereq)
    print("")
    
#same_courses_data = get_course_subsections("angularjs for beginners")
#for c in same_courses_data:
    #for prereq in c:
        #print(prereq)
    #print("")   
```

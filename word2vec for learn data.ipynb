```python
import nltk
from nltk.tokenize import word_tokenize
from collections import Counter


def get_content_words(content):
    content_words_counter = Counter()
    for subcont in content["c"]:
        content_words_counter.update(word_tokenize(subcont["t"]))
    return content_words_counter
```
